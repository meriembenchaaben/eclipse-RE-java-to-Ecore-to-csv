##BM25

```python
file1= open('gdrive/My Drive/RI/topics.1-50.txt', "r")

configFile= { "casefold" : True,

"requested" : 1000,
 "queries" : [
 ]
}
             
# parse an xml file by name
jsonFile={}
i=0

for x in file1:
      
      if x.startswith('<title> Topic:'): 
          i=i+1
          x=x.split('<title> Topic:')[1]  
          x=x.split('\n')[0]
          x = x.replace('\'', '')
          x = x.replace('.', '')
          previous=configFile["queries"]
      
        
          j="{ \'number\' : \'" + str(i) + "\',  \'text\' : \' #stopword(#bm25(" + x + ")) \' }"
          print(j)
        
          y= eval(j )
    
          previous.append(y )
          
          configFile["queries"]=previous
          

configFile

with open('gdrive/My Drive/RI/data1.json', 'w') as f:
    json.dump(configFile, f)
```
